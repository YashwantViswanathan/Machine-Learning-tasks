```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('smoking_drinking.csv')

print(df.head(10))
print(df.describe())
print(df.columns)
print(df.info)
print(df.dtypes)
```

```python
#print(df.columns)

#Checking for missing values 
print(df.isnull().sum())
#Concluded no missing values

print(df['sex'].dtype)
print(df['DRK_YN'].dtype)
```

```python
#Finding categorical values
categorical = [var for var in df.columns if df[var].dtype == 'object']
print("There are {} categorical variables\n".format(len(categorical)))
print("The categorical variables are :\n\n", categorical)
#Concluded 2 categorical variables
```

```python
#Converting the data to numericals, to create better feasibility of analysis
df['DRK_YN'] = df['DRK_YN'].map({'Y': 1, 'N': 0, 'unknown': 0})
df['sex'] = df['sex'].map({'Male':1, 'Female':0})
```

```python
#Exploring numerical variables
numerical =  [var for var in df.columns if df[var].dtype!='object']
print('There are {} numerical variables\n'.format(len(numerical)))
print('The numerical variables are :\n\n', numerical)
#Concluded 22 columns numerical
```

```python
#Analysing and exploring the numerical variables
#Viewing distribution of age variable
#f, ax = plt.subplots(figsize = (10,8))
x = df['age']
ax = sns.displot(x, bins = 10, color = 'red')
plt.show()
```

```python
#Detecting outliers in age variable using box plots
#f, ax = plt.subplots(figsize = (10,8))
x = df['age']
ax = sns.boxplot(x)

plt.show()
```

```python
#Viewing frequency distribution of smoking variable

print(df['SMK_stat_type_cd'].unique())
print(df['SMK_stat_type_cd'].value_counts())
```

```python
#Analysing frequency distribution of snoking variable

f, ax = plt.subplots(figsize=(8, 6))
ax = sns.countplot(y="SMK_stat_type_cd", data=df, palette="Set1")
ax.set_title("Frequency distribution of smoking variable")
plt.show()
```

```python
#Analysing the smoking variable wrt sex 

f, ax = plt.subplots(figsize=(10, 8))
ax = sns.countplot(x="SMK_stat_type_cd", hue="sex", data=df, palette="Set1")
ax.set_title("Frequency distribution of smoking variable wrt sex")
plt.show()
```

```python
#We can establish relations between the columns weight, height and waistline

f, ax = plt.subplots(figsize=(10, 8))
ax = sns.countplot(x="height", hue="weight", data=df, palette="Set1")
ax.set_title("Frequency distribution of height variable wrt weight")
plt.show()
```
